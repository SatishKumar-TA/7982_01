```python
import numpy as np
import pandas as pd
import matplotlib
```

```python
df=pd.read_csv('C:/Users/satish.bogala/Documents/movie_metadata.csv',escapechar="\\")
```

## Directors with highest facebook likes

```python
df1=df.groupby('director_name')['director_facebook_likes'].sum().reset_index()
df1.sort_values(by="director_facebook_likes",inplace=True,ascending=False)
df1
```

## No of movies released in a particular language in a year along with budget

```python
df2=df.groupby(['title_year','language']).agg({'movie_title':pd.Series.count,'budget':np.sum}).reset_index()
df2
```

## Graph between year and total budget

```python
df2[20:].plot(x='title_year', y='budget', figsize=(10,5), grid=True , color='DarkBlue')
```

## Graph between langauges and its Max gross

```python
df3=df.groupby('language')['gross'].max().reset_index()
df4=df3.sort_values(by="gross",ascending=False)
df3.plot(x='language', y='gross', figsize=(10,5), grid=True , color='DarkBlue')
```
